```python
import dataiku
from dataiku import pandasutils as pdu
import pandas as pd
import numpy as np
```

```python
# Example: load a DSS dataset as a Pandas dataframe
mydataset = dataiku.Dataset("Objects_dvc_by_WEGVAK")
df = mydataset.get_dataframe()
```

```python
df.head()
```

```python
#calculate dvc after ecopassage
df['DVC_after'] = df['DVC_first'] - df['DVC_before_first']
```

```python
#df before ecopassage with nan distance to ecopassage
df_voor = df.copy()
df_voor['ecopassage'] = 0
df_voor['Afstand_zeebrug_x_first'] = np.nan
df_voor['DVC'] = df_voor['DVC_before_first']
```

```python
#df after ecopassage
df_na = df.copy()
df_na['ecopassage'] = 1
df_na['DVC'] = df_na['DVC_after']
```

```python
df_na.head()
```

```python
#Make df with road segments before and after ecopasage 
model_input = pd.concat([df_na, df_voor])
```

```python
model_input.head()
```

```python
# Recipe outputs
model = dataiku.Dataset("Model_input")
model.write_with_schema(model_input)
```
