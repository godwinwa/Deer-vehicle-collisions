```python
# Example: load a DSS dataset as a Pandas dataframe
mydataset = dataiku.Dataset("Zeeweg_Night_Day")
df = mydataset.get_dataframe()
```

```python
df.head()
```

```python
#change column from string to datetime 
df['Datum']= pd.to_datetime(df['Datum'])
```

```python
#get weekday
def get_weekday(day):
    return day.strftime("%w")

df['Day'] = df['Datum'].apply(get_weekday)
```

```python
#weekend or workday
def weekend(day):
    if day == '5' or day == '6':
        return 'weekend'
    return 'workday'

df['Weekend'] = df['Day'].apply(weekend)
```

```python
#get hour
df['hour']=df['Tijd'].apply(lambda x: x[:2])
```

```python
#get hour
hour = df.hour
```

```python
#remove 00:00 time
hour = hour.value_counts()[1:]
```

```python
#distribution hour
df.Tijd.value_counts()
```

```python
#built in chisquare test
chisquare(hour)
```

```python
#function to get chisquare
def pearsonChiScuare(df, exp=None):


    group = df.Day
    group = group.value_counts()


    df = len(group) - 1

    if exp ==None:
        minE = sum(group)/ len(group)
        chiVal, pval = chisquare(group)
    else:
        minE = min(exp)
        chiVal, pval = chisquare(group, exp)

    warning = None
    if minE < 5:
        warning = 'Minimum expected count less than 5'

    return chiVal, pval, df, minE, warning
```

```python
pearsonChiScuare(df)
```

```python
#get relative numbers of weekend/workday accidents by dividing by traffic intensity 
#store in dictionary 
traffic_intensity = {2014: [12344, 11779], 2015: [12187, 11407],
                     2016: [11964, 11185], 2017: [12158, 11490],
                     2018: [12672, 12079], 2019: [12164, 11154],
                     2020: [11191, 10992]}

weekend_relative = {}
workday_relative = {}

for i in df['Year'].unique():
    weekend_relative[i] = len(df[(df['Year'] == i) & (df['Weekend'] == 'weekend')]) / traffic_intensity[i][0] * 11000

    workday_relative[i] = len(df[(df['Year'] == i) & (df['Weekend'] == 'workday')]) / traffic_intensity[i][1] * 11000
```

```python
#create df of weekend/workday 
df_week_workday = pd.DataFrame([weekend_relative, workday_relative])
```

```python
#transpose dataframe
df1 = df_week_workday.T
```

```python
#rename columns
df1 = df1.rename(columns={0:"Weekend", 1:'Workday'})
```

```python
#get weekday total
weekend = df1['Weekend'].sum()
workday = df1['Workday'].sum()
```

```python
chisquare([weekend, workday])
```
