```python
import dataiku
from dataiku import pandasutils as pdu
```

```python
import pandas as pd
from astral import Astral
import datetime as datetime
import dateutil, pytz
```

```python
# Example: load a DSS dataset as a Pandas dataframe
mydataset = dataiku.Dataset("Zeeweg_DVCs")
df = mydataset.get_dataframe()
```

```python
df.head()
```

```python
#change column to datetime
df['Datum']= pd.to_datetime(df['Datum'])
df['Tijd'] = pd.to_datetime(df['Tijd'],format= '%H:%M' ).dt.time
```

```python
#change column time to string 
df['Tijd'] = (df['Tijd']).astype(str)
```

```python
# check if time between sunrise and sunset
#store in dictionary
daylight = {}

city_name = 'Amsterdam'
a = Astral()
a.solar_depression = 'civil'
city = a[city_name]
for i in range(0, 171):
    date = df.iloc[i]['Datum']
    time = date.strftime("%Y-%m-%d")
    T = datetime.datetime.combine(datetime.date(int(time[:4]),
        int(time[5:7]), int(time[8:10])),
        datetime.time(int(df.iloc[i]['Tijd'][:2]), int(df.iloc[i]['Tijd'][3:5])))
    sun = city.sun(date=datetime.date(int(time[:4]),
        int(time[5:7]), int(time[8:10])), local=True)
    now = pytz.utc.localize(T)
    if (sun['sunrise'] < now < sun['sunset']):
        daylight[i] = 'Y'
    else:
        daylight[i] = 'N'
```

```python
df.head()
```

```python
daylight
```

```python
#add daylight dict to df
df['Daylight'] = df.index.to_series().map(daylight)
```

```python
# Recipe outputs
zeeweg_dvc = dataiku.Dataset("Zeeweg_Night_Day")
zeeweg_dvc.write_with_schema(df)
```
